```python
import datetime

show_daily_cash_range_per_region = False
show_daily_cash_per_monster_per_region = False
show_cost_of_upgrades_per_level = False
show_kill_times_per_monster_per_region = False
show_different_gold_from_monsters_per_region = False
show_time_for_craftables_range_per_region = False
show_craft_times_per_item_per_region = False
show_max_items_per_item_per_day = False
show_gold_from_crafting_items = False
show_xp_from_crafting_items = False
show_renown_per_region = False


def print_out_results(double_list, second_quantity):
    for tier in range(1, 11):
        value = str(tier)

        for i in range(1,second_quantity+1):
            value += '\t' + str(double_list[tier-1][i-1])

        print(value)

        
def get_time(seconds):
    min, sec = divmod(seconds, 60)
    hour, min = divmod(min, 60)
    
    if hour == 0 and min == 0:
        return "%ds" % (sec)
    elif hour == 0:
        return "%dm%02ds" % (min, sec)
    else:
        return "%dh%02dm" % (hour, min)

#================================
# Daily Cash Range Per Region
#================================
if show_daily_cash_range_per_region:
    print('\n==================================')
    print('\n Daily Cash Range Per Region \n')
    print('==================================\n')

start_cash = 4000
multiplier_per_tier = 1.2

min_ranges = []
total_ranges = []

for i in range(1,11):
    previous_avg = 0
    
    if i == 1:
        previous_avg = start_cash
    else:
        previous_avg = ((i-1) * start_cash) ** multiplier_per_tier
    
    current_avg = (i * start_cash) ** multiplier_per_tier
    next_avg = ((i+1) * start_cash) ** multiplier_per_tier
    
    min_range = round(current_avg - (current_avg - previous_avg) / 2)
    max_range = round(current_avg + (next_avg - current_avg) / 2)
    total_range = round(max_range - min_range)
    
    min_ranges.append(min_range)
    total_ranges.append(total_range)
    
    #==FULL DETAILS
    #print('Tier ' + str(i) + ': Range = ' + str(min_range) + ' - ' + str(max_range) + ' || (Range total = ' + str(total_range) + ')')
    
    #==EXPORT
    if show_daily_cash_range_per_region:
        print(str(i) + '\t' + str(min_range) + '\t' + str(max_range) + '\t' + str(total_range))

    
#================================
# Daily Cash Per Monster Per Region
#================================
if show_daily_cash_per_monster_per_region:
    print('\n==================================')
    print('\n Daily Cash Per Monster Per Region \n')
    print('==================================\n')
    
cash_per_monster_per_region = []
num_of_monsters = 8
percent_added_per_monster = (100/8) / 100

for tier in range(1,11):
    cash_per_monster = []
    
    for monster in range(1,num_of_monsters+1):
        cash_per_day = min_ranges[tier-1] + round(total_ranges[tier-1] * (percent_added_per_monster * monster))
        cash_per_monster.append(cash_per_day)
        
    cash_per_monster_per_region.append(cash_per_monster)

if show_daily_cash_per_monster_per_region:
    print_out_results(cash_per_monster_per_region, num_of_monsters)
    
    
#================================
# Cost of Upgrades Per Level
#================================
if show_cost_of_upgrades_per_level:
    print('\n==================================')
    print('\n Cost of Upgrades Per Level \n')
    print('==================================\n')

bonus_per_level = 2
cost_per_level = 200
cost_multiplier_per_level = 1.15

for level in range(1,16):
    cost_to_level = round((cost_per_level * level) * cost_multiplier_per_level)
    bonus = bonus_per_level * level
    
    if show_cost_of_upgrades_per_level:
        print('Level ' + str(level) + ' costs = ' + str(cost_to_level) + ' || Bonus = ' + str(bonus) + '%')


#================================
# Kill Times Per Monster, Per Region
#================================
if show_kill_times_per_monster_per_region:
    print('\n==================================')
    print('\n Kill Times Per Monster, Per Region \n')
    print('==================================\n')

seconds_per_kill = 50
tier_multiplier = 1.2

seconds_to_kill_per_region = []
kills_per_day_per_region = []

#The region tiers
for tier in range(1, 11):
    seconds_to_kill = []
    kills_per_day = []

    #The monsters per tier
    for i in range(1,9):
        seconds = round(((seconds_per_kill + (5 * i))) ** (tier_multiplier + (tier / 100)))
        seconds_to_kill.append(seconds)
        
        kills = round((24 * 60 * 60) / seconds)
        kills_per_day.append(kills)

    seconds_to_kill_per_region.append(seconds_to_kill)
    kills_per_day_per_region.append(kills_per_day)
        
if show_kill_times_per_monster_per_region:
    print('== All Kill Times Per Tier\n')
    for tier in range(1, 11):
        list_of_times = seconds_to_kill_per_region[tier-1]

        all_kill_times = str(tier)

        for i in range(1,9):
            seconds = list_of_times[i-1]
            all_kill_times += '\t' + str(seconds) + '\t' + str(seconds * 80) + '\t' + get_time(seconds * 80)

        print(all_kill_times)

if show_kill_times_per_monster_per_region:
    print('\n== Max Kills Per Day Per Tier\n')
    print_out_results(kills_per_day_per_region, 8)

    
#================================
# Gold Drops, Gold From Loot, Gold from Trinkets - Per Monster, Per Region
#================================
if show_different_gold_from_monsters_per_region:
    print('\n==================================')
    print('\n Gold Drops, Gold From Loot, Gold from Trinkets - Per Monster, Per Region \n')
    print('==================================\n')

p_gold_from_monsters = 0.2
p_gold_from_gear_drops = 0.12


def get_gold_from_percent_of_daily(percentage):
    gold_per_monster_per_region = []
    
    for tier in range(1, 11):
        gold_per_monster = []
        
        for i in range(1,9):
            cash_per_monster = cash_per_monster_per_region[tier-1][i-1]
            cash_with_percentage = round(cash_per_monster * percentage)
            
            cash_per_kill = round(cash_with_percentage / kills_per_day_per_region[tier-1][i-1])
    
            gold_per_monster.append(cash_per_kill)
    
        gold_per_monster_per_region.append(gold_per_monster)
    
    return gold_per_monster_per_region

if show_different_gold_from_monsters_per_region:
    print('== Gold from monster kills - per monster, per region\n')
    gold_from_monsters = get_gold_from_percent_of_daily(p_gold_from_monsters)
    print_out_results(gold_from_monsters, 8)

    print('\n== Gold from monster drops - per monster, per region\n')
    gold_from_gear_drops = get_gold_from_percent_of_daily(p_gold_from_gear_drops)
    print_out_results(gold_from_gear_drops, 8)


#================================
# Time for craftables - Range Per Region
#================================
if show_time_for_craftables_range_per_region:
    print('\n==================================')
    print('\n Time for craftables - Range Per Region\n')
    print('==================================\n')

p_gold_from_monsters = 0.2

start_time = 200
upfront_added_seconds = 482
multiplier_per_tier = 1.1

min_time_ranges = []
total_time_ranges = []

for i in range(1,11):
    previous_avg = 0
    
    if i == 1:
        previous_avg = ((i / 2) * (start_time / 1.5)) ** multiplier_per_tier + upfront_added_seconds
    else:
        previous_avg = (((i-1) / 2) * (start_time)) ** multiplier_per_tier + upfront_added_seconds
    
    current_avg = (i * start_time) ** multiplier_per_tier + upfront_added_seconds
    next_avg = (((i+1) / 2) * start_time) ** multiplier_per_tier + upfront_added_seconds
    
    min_range = round(current_avg - (current_avg - previous_avg))
    max_range = round(current_avg + (next_avg - current_avg))
    total_range = round(max_range - min_range)
    
    min_time_ranges.append(min_range)
    total_time_ranges.append(total_range)
    
    #==FULL DETAILS
    items_per_day_min = round((60 * 60 * 24) / min_range)
    items_per_day_max = round((60 * 60 * 24) / max_range)
    
    #==EXPORT
    if show_time_for_craftables_range_per_region:
        #print('Tier ' + str(i) + ': Range = ' + get_time(min_range) + ' (' + str(items_per_day_min) + ') - ' + get_time(max_range) + ' (' + str(items_per_day_max) + ')')
        print(str(i) + '\t' + str(min_range) + '\t' + str(max_range) + '\t' + get_time(min_range) + '\t' + get_time(max_range) + '\t' + str(items_per_day_min) + '\t' + str(items_per_day_max))
    
    
#================================
# Craft Times Per Item Per Region
#================================
if show_craft_times_per_item_per_region:
    print('\n==================================')
    print('\n Craft Times Per Item Per Region \n')
    print('==================================\n')

craft_times_per_item_per_region = []
num_of_items = 14
percent_added_per_item = (100/num_of_items) / 100

for tier in range(1,11):
    cash_per_item = []
    
    for item in range(1,num_of_items+1):
        cash_per_day = min_time_ranges[tier-1] + round(total_time_ranges[tier-1] * (percent_added_per_item * item))
        cash_per_item.append(cash_per_day)
        
    craft_times_per_item_per_region.append(cash_per_item)

if show_craft_times_per_item_per_region:
    print_out_results(craft_times_per_item_per_region, num_of_items)
        

#================================
# Max Items Per Item Per Day Per Region
#================================
if show_max_items_per_item_per_day:
    print('\n==================================')
    print('\n Max Items Per Item Per Day \n')
    print('==================================\n')

max_items_per_day_per_region = []

#The region tiers
for tier in range(1, 11):
    items_per_day = []

    #The items per tier
    for i in range(1,num_of_items+1):
        seconds = craft_times_per_item_per_region[tier-1][i-1]

        max_per_day = round((24 * 60 * 60) / seconds)
        items_per_day.append(max_per_day)

    max_items_per_day_per_region.append(items_per_day)
        
if show_max_items_per_item_per_day:
    print_out_results(max_items_per_day_per_region, num_of_items)

    
#================================
# Gold from Selling, Gold used to Craft - Per Item, Per Region
#================================

materials_for_each_recipe = [1,1,2,2,3,3,3,4,4,4,3.5,6,6,10]
p_added_bonus_per_item_type = 0.5

if show_gold_from_crafting_items:
    print('\n==================================')
    print('\n Gold from Selling, Gold used to Craft - Per Item, Per Region \n')
    print('==================================\n')

def get_gold_from_items_from_percent_of_daily(percentage):
    gold_per_item_per_region = []
    
    for tier in range(1, 11):
        gold_per_item = []
        
        count = 0
        
        for i in range(1,num_of_items+1):
            daily_cash = cash_per_monster_per_region[tier-1][count]
            max_items_per_day = max_items_per_day_per_region[tier-1][i-1]
            
            if i % 2 == 0:
                count += 1
            
            cash_with_percentage = round(daily_cash * percentage)
            
            cash_per_item = cash_with_percentage / (max_items_per_day * materials_for_each_recipe[i-1])
            cash_based_on_materials = cash_per_item * materials_for_each_recipe[i-1]
            cash_with_extra_bonus = cash_based_on_materials + (cash_based_on_materials * (i * p_added_bonus_per_item_type))
    
            gold_per_item.append(round(cash_with_extra_bonus))
    
        gold_per_item_per_region.append(gold_per_item)
    
    return gold_per_item_per_region


if show_gold_from_crafting_items:
    print('\n== Gold from Selling Craftables - per item, per region\n')
    p_gold_from_selling_items = 0.3

    output_list = get_gold_from_items_from_percent_of_daily(p_gold_from_selling_items)
    print_out_results(output_list, num_of_items)

    
if show_gold_from_crafting_items:
    print('\n== Gold used to Craft Items - per item, per region\n')
    p_gold_used_to_craft_items = 0.12

    output_list = get_gold_from_items_from_percent_of_daily(p_gold_used_to_craft_items)
    print_out_results(output_list, num_of_items)


#================================
# XP Per Crafting Level
#================================
if show_xp_from_crafting_items:
    print('\n==================================')
    print('\n Gold from Selling, Gold used to Craft - Per Item, Per Region \n')
    print('==================================\n')
    
days_to_beat_tiers = [0,2,3,5,7,9,12,16,19,24,28]

start_exp = 100
tier_xp_multiplier = 1.3

def get_xp_for_level(level):
    return round(((level) * start_exp) ** tier_xp_multiplier + start_exp)

for i in range(1, num_of_levels+1):
    exp_for_level = get_xp_for_level(i)
    #print(str(i) + '\t' + str(exp_for_level))

import math

xp_per_item_per_region = []

for tier in range(1, 11): 
    xp_to_get_tier = 0
    
    days_for_tier = days_to_beat_tiers[tier]
    days_in_seconds = days_for_tier * (60 * 60 * 24)

    #Get total xp for the tier/region for the 14 item group
    for i in range(1, num_of_items+1):
        xp_to_get_tier += get_xp_for_level(tier * i)
    
    xp_for_tier_items = []
    
    for i in range(1, num_of_items+1):
        xp_for_level = get_xp_for_level(tier * i)
        level_xp_weight_of_total = xp_for_level / xp_to_get_tier
        
        seconds_for_level = days_in_seconds * level_xp_weight_of_total
        
        num_items_to_level = seconds_for_level / craft_times_per_item_per_region[tier-1][i-1]
        xp_per_item = math.ceil(xp_for_level / (num_items_to_level * materials_for_each_recipe[i-1]))
        xp_based_on_materials = xp_per_item * materials_for_each_recipe[i-1]
        xp_with_extra_bonus = xp_based_on_materials + (xp_based_on_materials * (i * p_added_bonus_per_item_type))
    
        xp_for_tier_items.append(round(xp_with_extra_bonus))
        
    xp_per_item_per_region.append(xp_for_tier_items)

if show_xp_from_crafting_items:
    print_out_results(xp_per_item_per_region, num_of_items)

    
#================================
# Renown Per Region
#================================
if show_renown_per_region:
    print('\n==================================')
    print('\n Renown Per Region \n')
    print('==================================\n')

start_renown = 150
tier_renown_multiplier = 1.8

def get_renown_for_tier(tier):
    return round(((tier) * start_renown) ** tier_renown_multiplier + start_renown)

for i in range(1, 11):
    renown_for_level = get_renown_for_tier(i)
    #print(str(i) + '\t' + str(renown_for_level))


#================================
# Renown Per Monster and Request Per Region
#================================
if show_renown_per_region:
    print('\n==================================')
    print('\n Renown Per Monster and Request Per Region \n')
    print('==================================\n')

daily_requests = 6
requests_weight = 0.2
monsters_weight = 0.8


renown_per_monster_per_region = []

for tier in range(1, 11):
    renown_per_region = []
    
    renown_to_get_tier = get_renown_for_tier(tier)
    renown_to_get_tier *= monsters_weight
    days_for_tier = days_to_beat_tiers[tier]
    
    monster_bonus = -0.2
    
    for i in range(1, 9):
        max_kills_per_day = kills_per_day_per_region[tier-1][i-1]
        kills_to_level = max_kills_per_day * days_for_tier

        avg_renown_per_kill = renown_to_get_tier / kills_to_level
        
        if i == 8:
            monster_bonus = 0.16
        else:
            monster_bonus += 0.04
        
        avg_renown_per_kill = avg_renown_per_kill + (avg_renown_per_kill * monster_bonus)

        renown_per_region.append(round(avg_renown_per_kill))
        
    renown_per_monster_per_region.append(renown_per_region)


if show_renown_per_region:
    print('\n== Renown Per Monster \n')
    print_out_results(renown_per_monster_per_region, 8)

if show_renown_per_region:
    print('\n== Renown Per Request Per Region \n')
    
    for tier in range(1, 11):
        renown_to_get_tier = get_renown_for_tier(tier)
        renown_to_get_tier *= requests_weight
        
        days_for_tier = days_to_beat_tiers[tier]
        
        requests_to_level = days_for_tier * daily_requests
        avg_renown_per_request = renown_to_get_tier / requests_to_level
        
        print(str(tier) + '\t' + str(round(avg_renown_per_request)))
        



```
